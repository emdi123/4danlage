```python
#########################
#List comprehension
#########################
#How do we comprehend a list in python

#We have to do the follwoing in python
text=" How are u doing 21"

word = []
for i in text:
    if i in ['2','1']:
         new = i+'Din'
         word.append(new)
print('the old word is'+ str(word))

#Now let's do in with the list comprehension method in python

words =[ i +'Din' for i in text if  i in ['2','1']]

print('The new word list is ' + str(words))




    
```
